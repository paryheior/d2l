```python
%matplotlib inline
import numpy as np
from mpl_toolkits import mplot3d
from d2l import torch as d2l
import torch
```

```python
def f(x):
    return x*torch.cos(np.pi*x)
```

```python
d2l.set_figsize((4.5,2.5))
```

```python
x = torch.arange(-1.0,2.0,0.01)
d2l.plot(x, [f(x), ], 'x', 'f(x)')
```
